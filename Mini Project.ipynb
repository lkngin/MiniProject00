```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
sns.set(color_codes=True)
```

Import Sephora Website Dataset

```python
colnames = ['category', 'brand', 'rating', 'number_of_reviews', 'love', 'price', 'value_price', 'ingredients']
dfSephora = pd.read_csv('sephora_website_dataset.csv', usecols = colnames)
```

```python
dfSephora
```

Find the information of the dataset

```python
dfSephora.info()
```

Top 10 brands highest rating and love then categories them

1. Sort the rating and love. Select top 10.

```python
dfTop = dfSephora.sort_values(by=['rating', 'love'], ascending=False).head(10)
```

```python
dfCB = dfTop.groupby(['brand'])
dfCB.first()
```

```python
import matplotlib.pyplot as plt
ax = dfTop[['love']].plot(kind='bar', title ="Brand Love", figsize=(15, 10), legend=True, fontsize=12)
ax.set_xlabel("Brand", fontsize=12)
ax.set_ylabel("Love", fontsize=12)
plt.show()
```

Summary
From this data, what can we see is that Anastasia Beverly Hills which is a Eye Palettes are the most well loved by the Sephora customers.
